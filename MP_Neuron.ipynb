#MP Neuron Class

```
class MPNeuron:
  def __init__(self):
    self.b = None
  
  def model(self, x):
    return (sum(x) >= self.b)
  # (sum(x) >= b) return True or False
  
  def predict(self, X):
    Y = []
    for x in X:
      result = self.model(x)
      Y.append(result)
    return np.array(Y)
  
  def fit(self, X, Y):
    accuracy = {}
    
    for b in range(X.shape[1] + 1):
      self.b = b
      Y_pred = self.predict(X)
      accuracy[b] = accuracy_score(Y_pred, Y)
      # gets max value of b from dictionary of accuracy
    best_b = max(accuracy, key = accuracy.get)
    self.b = best_b
    
    print ("Optimal b", best_b)
    print("highest accuracy ",accuracy[best_b])
```
